```python
import yfinance as yf
import pandas as pd

# List of Nifty 50 tickers
nifty_50_tickers = [
    'ADANIENT.NS', 'ADANIPORTS.NS', 'APOLLOHOSP.NS', 'ASIANPAINT.NS', 'AXISBANK.NS', 
    'BAJAJ-AUTO.NS', 'BAJFINANCE.NS', 'BAJAJFINSV.NS', 'BEL.NS', 'BPCL.NS', 
    'BHARTIARTL.NS', 'BRITANNIA.NS', 'CIPLA.NS', 'COALINDIA.NS', 'DRREDDY.NS', 
    'EICHERMOT.NS', 'GRASIM.NS', 'HCLTECH.NS', 'HDFCBANK.NS', 'HDFCLIFE.NS', 
    'HEROMOTOCO.NS', 'HINDALCO.NS', 'HINDUNILVR.NS', 'ICICIBANK.NS', 'ITC.NS', 
    'INDUSINDBK.NS', 'INFY.NS', 'JSWSTEEL.NS', 'KOTAKBANK.NS', 'LT.NS', 
    'M&M.NS', 'MARUTI.NS', 'NTPC.NS', 'NESTLEIND.NS', 'ONGC.NS', 
    'POWERGRID.NS', 'RELIANCE.NS', 'SBILIFE.NS', 'SHRIRAMFIN.NS', 'SBIN.NS', 
    'SUNPHARMA.NS', 'TCS.NS', 'TATACONSUM.NS', 'TATAMOTORS.NS', 'TATASTEEL.NS', 
    'TECHM.NS', 'TITAN.NS', 'TRENT.NS', 'ULTRACEMCO.NS', 'WIPRO.NS'
]

start_date = "2013-01-01"
end_date = "2023-01-01"

all_data = []

for ticker in nifty_50_tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    
    # Ensure data is non-empty and has required columns
    if not data.empty:
        data['Symbol'] = ticker
        data.reset_index(inplace=True)  # Make Date a column

        # Check if required columns exist in the data
        for column in ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']:
            if column not in data.columns:
                data[column] = pd.NA

        # Debug print: print a sample of the data to confirm structure
        print(f"Sample data for {ticker}:\n", data.head())
        
        # Add data to the list
        all_data.append(data)
    else:
        print(f"No data for {ticker}")

# Combine all data
if all_data:
    combined_data = pd.concat(all_data, ignore_index=True)
    print("Data concatenated successfully.")
    
    # Debug print: Show the columns of combined_data
    print("Columns in combined_data:", combined_data.columns)
    print(combined_data.head())
else:
    raise ValueError("No data downloaded for any tickers.")

# Check if required columns are in the combined DataFrame
required_columns = {'Date', 'Symbol', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume'}
missing_columns = required_columns - set(combined_data.columns)
if missing_columns:
    raise KeyError(f"The following required columns are missing in combined_data: {missing_columns}")

# Proceed with data transformation if columns are as expected
combined_series = combined_data[['Date', 'Symbol', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']]
combined_series = combined_series.melt(id_vars=['Date', 'Symbol', 'Volume'], var_name='Metric', value_name='Value')

# Save to CSV
combined_series.to_csv('stock_data_series.csv', index=False)
print("Data downloaded and saved to 'stock_data_series.csv'.")
```
